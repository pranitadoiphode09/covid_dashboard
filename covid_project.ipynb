```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("covid_daily_full.csv")

icu_df = pd.read_csv("covid_daily_hospital_icu.csv")

testing_df = pd.read_csv("covid_daily_testing.csv")

vaccination_df = pd.read_csv("covid_daily_vaccinations.csv")
```

```python
continent_lookup = df[['location', 'continent']].drop_duplicates()

icu_df = icu_df.merge(continent_lookup, on='location', how='left')

testing_df = testing_df.merge(continent_lookup, on='location', how='left')

vaccination_df = vaccination_df.merge(continent_lookup, on='location', how='left')
```

```python

cases_continent = df.groupby(['continent', 'date']).agg({
    'new_cases_smoothed': 'sum',
    'new_deaths_smoothed': 'sum'
}).reset_index()

icu_continent = icu_df.groupby(['continent', 'date']).agg({
    'icu_patients_per_million': 'mean'
}).reset_index()

testing_continent = testing_df.groupby(['continent', 'date']).agg({
    'positive_rate': 'mean'
}).reset_index()

vaccination_continent = vaccination_df.groupby(['continent', 'date']).agg({
    'people_fully_vaccinated_per_hundred': 'mean'
}).reset_index()
```

```python
print(df.columns) 
print()
print(icu_df.columns)
print()
print(testing_df.columns)
print()
print(vaccination_df.columns)
```

```python
df.fillna(0, inplace=True)
print(df.isnull().sum())

icu_df.fillna(0, inplace=True)
print(icu_df.isnull().sum())

testing_df.fillna(0, inplace=True)
print(testing_df.isnull().sum())

vaccination_df.fillna(0, inplace=True)
print(vaccination_df.isnull().sum())
```

```python
df["date"] = pd.to_datetime(df["date"],format="mixed",dayfirst=True)
icu_df['date'] = pd.to_datetime(icu_df['date'],format="mixed",dayfirst=True)
testing_df['date'] = pd.to_datetime(testing_df['date'],format="mixed",dayfirst=True)
vaccination_df['date'] = pd.to_datetime(vaccination_df['date'],format="mixed",dayfirst=True)
```

```python
df["continent"] = df["continent"].replace(0, np.nan)
df = df.dropna(subset=["continent"])
```

```python

total_cases_plot = df.groupby("continent")["total_cases"].max()

plt.figure(figsize=(10,7))

plt.pie(total_cases_plot.values, 
        labels=total_cases_plot.index,
        autopct = "%1.1f%%",
        startangle=140)

plt.title("Total COVID-19 Cases by Continent", fontsize=15,pad=55)
plt.axis('equal')
plt.show()
```

```python

monthly_cases = (df.set_index('date').resample('ME')['new_cases_smoothed'].sum().reset_index())
sns.set_style("whitegrid")

plt.figure(figsize=(10, 7))
sns.lineplot(data= monthly_cases, x='date',y='new_cases_smoothed')

plt.title('Monthly COVID-19 Cases Trend', fontsize=15,pad=20)
plt.xlabel('Year')
plt.ylabel('New Cases')
plt.show()
```

```python
monthly_deaths = (df.set_index('date').resample('ME')['new_deaths_smoothed'].sum() )

plt.figure(figsize=(10,7))
plt.plot(monthly_deaths.index, monthly_deaths.values)

plt.xlabel("Year")
plt.ylabel("New Cases")
plt.title("Monthly COVID-19 Cases Trend", fontsize=15,pad=20)
plt.grid(True)
plt.show()
```

```python
cases_by_location = (df.groupby("location")["total_cases_per_million"].max().sort_values(ascending=False).head(10).reset_index())

plt.figure(figsize=(9,5))

sns.barplot(
    data = cases_by_location,
    x = "location",
    y = "total_cases_per_million",
    palette="viridis",hue="location"
)

plt.xlabel("Country")
plt.ylabel("Total Cases per Million")
plt.title("Top 10 Countries by Total Cases per Million", fontsize=15,pad=20)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python

deaths_by_location = (df.groupby("location")["total_deaths_per_million"].max().sort_values(ascending=False).head(10).reset_index())

plt.figure(figsize=(9,5))
sns.barplot(
    data=deaths_by_location,
    x="location",
    y="total_deaths_per_million",
    palette="magma",hue="location"
)

plt.xlabel("Country")
plt.ylabel("Total Deaths per Million")
plt.title("Top 10 Countries by Total Deaths per Million", fontsize=15,pad=20)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
icu_by_location = (df.groupby("continent")["icu_patients_per_million"].max().sort_values(ascending=False).head(10).reset_index())

plt.figure(figsize=(9,5))

sns.barplot(
    data = icu_by_location,
    x = "continent",
    y = "icu_patients_per_million",
    palette="cividis",hue="continent"
)

plt.xlabel("Country")
plt.ylabel("Total Cases per Million")
plt.title("Top 10 Countries by Total Cases per Million", fontsize=15,pad=20)
plt.xticks(rotation=45, ha="right")
plt.tight_layout()
plt.show()
```

```python
cd = cases_continent.groupby('continent')[["new_cases_smoothed","new_deaths_smoothed"]].sum().reset_index()

plt.figure(figsize=(9,6))
sns.scatterplot(data=cd,
                x='new_cases_smoothed',
                y='new_deaths_smoothed',
                hue='continent')

plt.xlabel('Total Smoothed COVID Cases')
plt.ylabel('Total Smoothed COVID Deaths')
plt.title('COVID Cases vs Deaths by Continent', fontsize=15,pad=20)

plt.show()
```

```python
positive_vs_death = df.groupby("continent")[["positive_rate","new_deaths_smoothed"]].sum()

plt.figure(figsize=(9,6))
sns.scatterplot(data=positive_vs_death, 
                x='positive_rate',
                y='new_deaths_smoothed', 
                 hue='continent')

plt.xlabel("Average Test Positivity Rate")
plt.ylabel("Total COVID-19 Deaths")
plt.title("Relation between COVID Positive rate and Total Deaths", fontsize=15,pad=20)
plt.show()
```

```python
tdtc = df.groupby('continent').agg({'total_deaths': 'max','total_cases': 'max'})

plt.figure(figsize=(9,6))
tdtc['death_rate'] = tdtc['total_deaths'] / tdtc['total_cases']

plt.figure(figsize=(9,6))
sns.barplot(data=tdtc, 
                x='continent',
                y='death_rate',
                palette="mako", 
                hue='continent')
plt.title("How Deadly Was COVID-19 Across Continents", fontsize=15,pad=20)
plt.xlabel("Continent")
plt.ylabel("Death Rate")
plt.show()
```
